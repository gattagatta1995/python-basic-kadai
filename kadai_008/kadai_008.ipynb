```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに0～45までのランダムな整数を代入する
var = random.randint(0,45)


# 変数varが3の倍数の場合はFizzを出力
if var % 3 == 0:
  print("Fizz")

# 変数varが5の倍数の場合はBizzを出力
if var % 5 == 0:
  print("Buzz")

# 変数varが３の倍数且つ5の倍数の場合はFizzBuzzを出力
elif var % 3 == 0 and var % 5 == 0:
  print("FizzBuzz")

# それ以外の場合は整数を出力
else:
  print(var)
```
