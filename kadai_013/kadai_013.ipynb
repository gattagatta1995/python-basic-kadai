```python
# 商品価格と消費税率から税込金額を計算して返す関数
def tax(price: int, tax_rate: int) -> str:

    # 消費税
    tax_amount = price * (tax_rate / 100)       # 消費税額

    # 税込価格
    total_price = price + tax_amount            # 税込み価格

    # 戻り値を出力
    return f"{int(total_price)} 円"


# 引数：110円 :10%
result = tax(110, 10)
print(result)
```
