```python
# for文の場合

array = ["水","金","地","火","木","土","天","海","冥"]

# enumerate関数を用いてインデックスを取り出し出力
for index, value in enumerate(array):
  print(f"{value}")

# while文の場合

array = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]

# インデックスがリストの長さに達するまで1回ループ
index = 0
while index < len(array):
    print(array[index])
    index += 1
```
