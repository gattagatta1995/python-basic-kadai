```python
# 商品価格と消費税率から税込金額を計算して返す関数
def tax(price: int, tax_rate: int) -> float:

    # 消費税額（小数点を保持）
    tax_amount = price * (tax_rate / 100)

    # 税込価格
    total_price = price + tax_amount

    # 数値のみで出力
    return total_price


# 引数：110円 :10%
result = tax(110, 10)
print(result)
```
