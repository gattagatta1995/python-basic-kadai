```python
# Humanクラスの定義
class Human:
    # コンストラクタ
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # メソッド
    def printinfo(self):
        print("名前：", self.name)
        print("年齢：", self.age)

# Humanクラスのインスタンス
human = Human("侍太郎", 30)

human.printinfo()
```
